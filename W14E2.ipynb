```python
"""Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo"""
df_sorted = df.sort_values(by='DealerPrice', ascending=False)
quartile= int(len(df_sorted["DealerPrice"])/4)
prezzo_minimo = df_sorted['DealerPrice'].head(quartile).min()
prezzo_massimo = df_sorted['DealerPrice'].head(quartile).max()
print("Il range di prezzo dei primi 25% dei prodotti più costosi è da", prezzo_minimo, "a", prezzo_massimo)

```

```python
"""Qual è il prezzo medio per i prodotti di colore blu?"""
prezzo_medio_blu = df[df['Color'] == "Blue"] ['DealerPrice'].mean()
round(prezzo_medio_blu,2)
```

```python
"""Qual è il prezzo medio per i prodotti di colore rosso o nero?"""
prezzo_medio_red_or_black = df[(df['Color'] == "Red")| (df['Color'] == "Black")] ['DealerPrice'].mean()
round(prezzo_medio_red_or_black,2)
```

```python
"""Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg?"""
prezzo_max_42_10kg = df[(df['Size'] == "42") & (df['Weight'] >10)] ['DealerPrice'].max()
round(prezzo_max_42_10kg,2) 
```

```python
"""Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento?"""
prodotti = df[(df['Size'] == "42") & (df['Weight'] >10) & (df['Color'] == "Silver")]   
risultato = prodotti[['EnglishProductName', 'StandardCost']]
risultato
```

```python
"""Selezioniamo i dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, >50"""
under_20 = df[df['Age (years)'] < 20]
between_20_30 = df[(df['Age (years)'] >= 20) & (df['Age (years)'] < 30)]
between_30_40 = df[(df['Age (years)'] >= 30) & (df['Age (years)'] < 40)]
between_40_50 = df[(df['Age (years)'] >= 40) & (df['Age (years)'] < 50)]
over_50 = df[df['Age (years)'] > 50]
print("<20:", under_20)
print("20-30:", between_20_30)
print("30-40:", between_30_40)
print("40-50:", between_40_50)
print(">50:", over_50)
```

```python
"""Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?"""
under_20 = df[df['Age (years)'] < 20] ['Diastolic blood pressure (mm Hg)'].mean()
between_20_30 = df[(df['Age (years)'] >= 20) & (df['Age (years)'] < 30)] ['Diastolic blood pressure (mm Hg)'].mean()
between_30_40 = df[(df['Age (years)'] >= 30) & (df['Age (years)'] < 40)] ['Diastolic blood pressure (mm Hg)'].mean()
between_40_50 = df[(df['Age (years)'] >= 40) & (df['Age (years)'] < 50)] ['Diastolic blood pressure (mm Hg)'].mean()
over_50 = df[df['Age (years)'] > 50] ['Diastolic blood pressure (mm Hg)'].mean()
print("diastolic_blood_pressure <20:", round(under_20,2))
print("diastolic_blood_pressure 20-30:", round(between_20_30,2))
print("diastolic_blood_pressure 30-40:", round(between_30_40,2))
print("diastolic_blood_pressure 40-50:", round(between_40_50,2))
print("diastolic_blood_pressure >50:", round(over_50,2))
```

```python
"""Quali sono le medie di charges rispetto a region? Ci sono differenze significative?"""
medie_spese_per_regione = df.groupby('region')['charges'].mean()
round(medie_spese_per_regione,2)
```

```python
"""E rispetto a smoker?"""
medie_spese_per_smoker = df.groupby('smoker')['charges'].mean()
round(medie_spese_per_smoker,2)
```

```python
""" E rispetto a sex?"""
medie_spese_per_sex = df.groupby('sex')['charges'].mean()
round(medie_spese_per_sex,2)
```

```python
"""Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?"""
df['bmi'].describe()
```

```python
df['bmi_quartile'] = pd.qcut(df['bmi'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
spese_mediche_per_quartile_bmi = df.groupby('bmi_quartile')['charges'].agg(['min', 'mean', 'max'])
print("\nMinimo, media e massimo delle spese mediche per ciascun quartile di BMI:")
print(round(spese_mediche_per_quartile_bmi,2))
```
