```python
"""Abbiamo un DataFrame di dipendenti:"""
employees_df = pd.DataFrame({ 'employee_id': [101, 102, 103, 104, 105], 'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'], 'department_id': [1, 2, 1, 2, 3] })
employees_df
```

```python
"""E un DataFrame di dipartimenti:"""
departments_df = pd.DataFrame({ 'department_id': [1, 2, 3], 'department_name': ['HR', 'IT', 'Finance'], 'location': ['New York', 'San Francisco', 'Chicago'] })
departments_df
```

```python
"""Unire questi DataFrame in base alla colonna comune department_id, in modo da avere nel risultato informazioni sia sui dipendenti che sui dipartimenti, usando la funzione .merge()"""
"""metedodo 2: merged_df = employees_df.merge(departments_df, how="inner", on="department_id")"""
merged_df = pd.merge(left=employees_df, right=departments_df, on='department_id')
merged_df
```

```python
"""Per ogni DataFrame, trasformare la colonna department_id nell'indice, facendo in modo che la modifica sia permanente"""
employees_df.set_index("department_id", inplace=True)
departments_df.set_index("department_id", inplace=True)
```

```python
"""poi unire i due dataset mediante il metodo .join()"""
joined_df = employees_df.join(departments_df)
joined_df
```

```python
"""Dal database AdventureWorksDW importiamo le tabelle dimemployee e dimemployeesalesterritory come DataFrame"""
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd
dotenv.load_dotenv(override=True)
username=os.getenv("username2")
password=os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)
show_tables = pd.read_sql("Show Tables", db_engine)
df_employee = pd.read_sql("SELECT * FROM dimemployee", db_engine)
df_employee_sales_territory = pd.read_sql("SELECT * FROM dimemployeesalesterritory", db_engine)
df_sales_territory = pd.read_sql("SELECT * FROM dimsalesterritory", db_engine)
db_engine.dispose()
show_tables
```

```python
"""Effettuiamo un join tra i due DataFrame usando le colonne EmployeeKey"""
df_employee.set_index("EmployeeKey", inplace=True)
df_employee_sales_territory.set_index("EmployeeKey", inplace=True)
joined_df = df_employee.join(df_employee_sales_territory)
joined_df
```

```python
joined_df.shape
```

```python
"""Importiamo ora la tabella dimsalesterritory"""
dimsalesterritory = pd.read_sql("SELECT * FROM dimsalesterritory", db_engine)
```

```python
"""ed effettuiamo un join tra questa e il DataFrame risultante della join precedente, usando le colonne SalesTerritoryKey"""
joined_df.set_index("SalesTerritoryKey", inplace=True)
dimsalesterritory.set_index("SalesTerritoryKey", inplace=True)
joined2_df = joined_df.join(dimsalesterritory)
joined2_df
```

```python
"""Su questo DataFrame contare quanti dipendenti ci sono per ogni paese (country) e per ogni regione (region)"""
conteggio_dipendenti = joined2_df.groupby(['SalesTerritoryCountry', 'SalesTerritoryRegion']).size()
conteggio_dipendenti
```

```python
"""Valutiamo la media del BaseRate per ogni paese: ci sono differenze?"""
round(joined2_df.groupby('SalesTerritoryCountry') ['BaseRate'].mean(),2)
```
