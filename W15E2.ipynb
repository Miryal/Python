```python
"""Esercizio Creiamo un DataFrame sintetico, che contiene i guadagni mensili di diverse annate, con il seguente codice:"""
years = 5 
a = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), 
                  "Anno": np.repeat(list(range(years)), 12), 
                  "Guadagni": np.random.randint(800, 5000, 12*years)})
a
```

```python
"""Calcola la somma cumulativa delle vendite per ogni mese utilizzando il metodo .cumsum()"""
a['Somma_cumulativa'] = a.groupby('Mese')['Guadagni'].cumsum()
a.sort_values('Mese')
```

```python
"""Come sopra, ma diviso per ogni anno"""
a['Guadagni_cumulativi_per_anno'] = a.groupby(['Anno'])['Guadagni'].cumsum()
a
```

```python
"""Dal database AdventureWorks estraiamo la tabella dimcustomer"""
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
dotenv.load_dotenv()
username=os.getenv("username2")
password=os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)
query = "SELECT * FROM dimcustomer"
df = pd.read_sql(query, db_engine)
db_engine.dispose()
df.shape
df
```

```python
"""Trasformiamo i nomi dei clienti in modo che abbiano solo lettere minuscole, e i cognomi in modo che abbiano solo lettere maiuscole"""
df['FirstName'] = df['FirstName'].str.lower()
df['LastName'] = df['LastName'].str.upper()
df
```

```python
"""Sulla colonna EmailAddress, utilizzando il metodo .str.split(), estraiamo nome utente e dominio"""
df[['NomeUtente', 'Dominio']] = df['EmailAddress'].str.split('@', expand=True)
df
```

```python
"""Sulla colonna Phone, estraiamo ogni parte del numero (ad es. da "1 (11) 500 555-0162" a ["1", "(11)", "500", "555-0162"])"""
df['Parts'] = df['Phone'].str.split()
print(df['Phone'])
```

```python
"""Utilizzando il metodo .str.contains(), estraiamo tutti gli indirizzi e-mail che contengono il numero 21"""
df.loc[df['EmailAddress'].str.contains('21'), ['EmailAddress']]
```

```python
"""Estraiamo tutti gli indirizzi e-mail che contengono il numero "20" oppure il numero 10"""
filtro_email = df[df['EmailAddress'].str.contains('10|20')]
filtro_email[['EmailAddress']]
```

```python
"""Calcolare la lunghezza di ogni indirizzo e-mail ed estrarre i dieci più lunghi e i dieci più corti"""
df['Length'] = df['EmailAddress'].str.len()
df_sorted = df.sort_values(by='Length')
dieci_più_corti = df_sorted.head(10)
dieci_più_lunghi = df_sorted.tail(10)
print(dieci_più_corti[['EmailAddress', 'Length']])
print(dieci_più_lunghi[['EmailAddress', 'Length']])
```

```python
"""Modificare il dominio degli indirizzi e-mail da "adventure-works.com" a "aw-db.com" mediante il metodo .str.replace()"""
df['EmailAddress'] = df['EmailAddress'].str.replace(r'@adventure-works\.com$', '@aw-db.com')
df['EmailAddress']
```

```python
"""Dalla colonna AddressLine1 estraiamo tutti gli indirizzi che contengono la sottostringa "Street" """
indirizzi_con_street = df[df['AddressLine1'].str.contains('Street', case=False)]
print(indirizzi_con_street)
```

```python
"""Dai beginner_datasets carichiamo in un DataFrame il file facebook.csv, che contiene dei post con data di pubblicazione, tipo (foto, video, …) e numero di reactions raccolte"""
df = pd.read_csv("C:\\Users\\Miriana Albergo\\EpicodeP\\facebook.csv")
df
```

```python
"""Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp"""
df['status_published'] = pd.to_datetime(df['status_published'])
df
```

```python
"""Utilizzando il metodo .df.strftime(), formatta le date delle pubblicazioni in un formato specifico (ad esempio, "%Y-%m-%d %H:%M:%S")."""
df['status_published'] = df['status_published'].dt.strftime("%Y-%m-%d %H:%M:%S")
df
```

```python
"""Utilizzando gli attributi .dt.year , .dt.month , .dt.day , .dt.dayofweek , .dt.dayofyear, ottieniamo informazioni specifiche sulle date delle transazioni, come l'anno, il mese, il giorno della settimana, il giorno dell'anno ecc"""
df['Anno'] = df['status_published'].dt.year
df['Mese'] = df['status_published'].dt.month
df['Giorno'] = df['status_published'].dt.day
df['Giorno_della_settimana'] = df['status_published'].dt.dayofweek 
df['Giorno_dell_anno'] = df['status_published'].dt.dayofyear
df
```

```python
"""Estraiamo solo i post relativi al 2012"""
df_2012 = df[df['status_published'].dt.year == 2012]
df_2012
```

```python
"""Estraiamo solo i post relativi a maggio 2018"""
df_maggio_2018 = df_maggio_2018 = df[(df['status_published'].dt.year == 2018) & (df['status_published'].dt.month == 5)]
df_maggio_2018
```

```python
"""Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana"""
num_post_weekend = len(df[df['status_published'].dt.dayofweek >= 5])
num_post_lavorativi = len(df[df['status_published'].dt.dayofweek < 5])
print(num_post_weekend)
print(num_post_lavorativi)
```

```python
"""Troviamo il primo e ultimo post pubblicati in ogni anno"""
df['Anno'] = df['status_published'].dt.year # Estrai l'anno da ciascuna data nella colonna "status_published"
primo_post_per_anno = df.groupby('Anno')['status_published'].min()
ultimo_post_per_anno = df.groupby('Anno')['status_published'].max()
print(primo_post_per_anno)
print(ultimo_post_per_anno)
```

```python
"""Utilizzando il metodo .tz_localize(), aggiungiamo un fuso orario alle date delle transazioni"""
df['status_published'] = df['status_published'].dt.tz_localize('UTC')
df
```

```python
"""Calcoliamo la differenza di tempo tra due transazioni"""
df = df.sort_values(by='status_published')
df['differenza_di_tempo_tra_transazioni'] = df['status_published'].diff()
df
```

```python
"""Quanti tipi di post ci sono? E quanti per ogni tipo?"""
conteggio_tipi_post = df['status_type'].value_counts()
num_tipi_post = len(conteggio_tipi_post)
print(num_tipi_post)
print(conteggio_tipi_post)
```
